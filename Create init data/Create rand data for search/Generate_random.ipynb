```python
from numpy import*
import numpy as np
import random
from matplotlib import pyplot as plt

import matplotlib.pyplot as plt
```

```python
Ranges = [[0,10],[-10,10],[-50,50],[-50,50]]
vector_size = 4
population = 1
```

```python
def p(t, vec):
    res = 0
    for i in range(len(vec)):
      res += vec[i]*np.sin(t*np.pi*(i+1)/82)/100#(vec[i] / (i + 1)) * t**(i+1)
    return res

def dp(t,vec):
    res = 0
    for i in range(len(vec)):
      res += vec[i] *(np.pi*(i+1)/82)* np.cos(t*np.pi*(i+1)/82)/100#vec[i] * t**i
    return res;

def newton_search(a,  b,  EPS,  vec):
    Xnn = 3 * EPS
    Xn = b
    X = EPS
    while (fabs(Xnn - X) > EPS):
        X = Xn;
        Xnn = Xn - ((p(Xn, vec) - 1.0) / dp(Xn, vec));
        Xn = Xnn;
    return Xnn;

def get_tEnd (Tmax,tau,EPS,vec):
  ans = 0
  t = 0
  while (t < Tmax):
      if (tau >= (Tmax - t)): tau = Tmax - t
      if ((1.0 - p(t, vec)) * (1.0 - p(t + tau, vec)) <= 0):
        ans = newton_search(t, t + tau, EPS, vec)
        if(ans < 20):
          return -1
        return ans
      if (p(t, vec) > p(t + tau, vec)):
        return -1
      t += tau
  return -1
```
